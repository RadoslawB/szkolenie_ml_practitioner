## Feel the hyperameters


### Balans: learning_rate (eta) vs n_estimators
To naczynia połączone.

n_estimators (Liczba drzew): Ile iteracji poprawiania błędów ma wykonać algorytm.

learning_rate (Krok nauki): hiperparametr kontrolujący szybkość, z jaką model uczy się na danych treningowych podczas procesu gradient boosting.

Zmniejszcie learning_rate (np. do 0.01), ale zobaczcie, co się stanie, jeśli nie zwiększycie n_estimators. Model będzie "niedouczony" (Underfitting).

Reguła kciuka: Niski learning_rate + wysokie n_estimators zazwyczaj daje najlepsze wyniki, ale kosztuje najwięcej czasu CPU.

### max_depth (Głębokość drzewa)
To jest Wasz suwak od Bias / Variance Tradeoff.
Utaw na  max_depth = 20 i ocen wyniki modelu train vs test
Co mają zrobić: Ustawcie max_depth = 20.

### Losowość (Regularyzacja): subsample i colsample_bytree
To zabezpieczenia przed tym, żeby drzewa nie były "kopiuj-wklej".

subsample (0.5 - 1.0): Jaki procent losowych wierszy widzi każde drzewo.
colsample_bytree (0.5 - 1.0): Jaki procent losowych kolumn (cech) widzi każde drzewo.

## Ewaluacja modelu (Train vs Test)

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

def evaluate(y_true, y_pred, label):
    mae = mean_absolute_error(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    print(f"{label} - MAE: {mae:.2f}, MAPE: {mape:.2%}")


# ROZWIĄZANIE:
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

evaluate(y_train, train_preds, "Wynik Treningowy")
evaluate(y_test, test_preds, "Wynik Testowy")
```
